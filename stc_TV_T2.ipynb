```python
# show the dataframe
fig = px.line(dataframe, y="Total_watch_time_in_hours")
fig.show()
```

```python
# Load the dataset
df=dataframe.copy()

# Display the first few rows of the dataframe to understand its structure
df.head()
```

```python
df.columns
```

```python
df['date_'] = pd.to_datetime(df['date_'])
df.set_index('date_', inplace=True)
```

```python
# Assuming your watch time data is in a column named 'Total_watch_time_in_hours'
model = auto_arima(df['Total_watch_time_in_hours'], seasonal=True, m=12, trace=True,
                   error_action='ignore', suppress_warnings=True)
```

```python
forecast_periods = 61  # May and June of 2018, assuming daily data
forecast = model.predict(n_periods=forecast_periods)
```

```python
forecast_index = pd.date_range(start='2018-05-01', periods=forecast_periods, freq='D')
forecast_df = pd.DataFrame(forecast, index=forecast_index, columns=['Total_watch_time_in_hours'])
```

```python
df.columns.tolist()
```

```python
df.index.name
```

```python
combined_df = pd.concat([df['Total_watch_time_in_hours'], forecast_df['Total_watch_time_in_hours']])
```

```python
combined_df = pd.concat([df, forecast_df])
combined_df.tail() # Check the last few entries to see the forecasted data
```

```python
# Ensure the index is recognized as a date by Plotly Express
combined_df.reset_index(inplace=True)
combined_df.rename(columns={'index': 'Date'}, inplace=True)

# Creating the line graph
fig = px.line(combined_df, x='Date', y='Total_watch_time_in_hours', title='Total Watch Time in Hours')

# Showing the graph
fig.show()
```
